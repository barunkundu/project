```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

Read data

```python
df = pd.read_csv('D:/work hard/Pandas/project/Engineering Student Journey/students.csv')
df.columns
```

Copy this data into a new variable 

```python
data = df.copy()
```

Drop Unnecessary columns which column's we dont't need in our analysis

```python
data.drop(columns=['Student ID', 'Name', 'Backlogs', 'Clubs', 'Attendance (%)', 'Alumni Path'], inplace=True)
data
```

See Information 

```python
data.info()
```

Handling Null-Values

```python
# See the sum of null value in each column
data.isnull().sum()
```

```python
# Remove rows which contain gather than 3 blank rows
data.dropna(thresh=data.shape[1] - 3, inplace=True)
data.info()
```

```python
# Fill Null value to "unassigned" for this two column
data['Internship Domain']=data['Internship Domain'].fillna('Unassigned')

data['Placement Domain']=data['Placement Domain'].fillna('Unassigned')
```

Split the variable "Skills"

```python
'''For safety purposes we make all skills into uppercase or lowercase. 
we make it lower case'''

skill_list = ['c++','python','java', 'sql', 'machine learning','web development', 'data science']

for i in skill_list:
    data[i]=data['Skills'].apply(lambda x: 'Yes' if i in str(x).lower() else 'No')

data
```

```python
# check 
data.isnull().sum()
```

1. Subject wise student distribution i) for all, ii) gender wise distribution

```python
fig, axes = plt.subplots(1,2, figsize = (10, 5))

# for axes[0]
b = data['Branch'].value_counts() # counts how many times each unique value appears in the Branch
percentages = (b / b.sum()) * 100 # count percentages 

plot1= b.plot(kind='bar',  color='skyblue', edgecolor='black', ax=axes[0]) # make plot

# add student percentage of each subject
for i, (count, pct) in enumerate(zip(b, percentages)):            
    plot1.text(i, count +0.5, f'{pct:.1f}%', ha='center', va='bottom', fontsize=10)

axes[0].set_xlabel("Subject")
axes[0].set_ylabel("count")
axes[0].set_title("Distribution of Students")

# for axes[1]
sns.countplot(x='Branch',hue='Gender', data=data ,ax=axes[1])
axes[1].set_xlabel("Subject")
axes[1].set_ylabel("count")
axes[1].set_title("Gender based Distribution of Students")

plt.tight_layout(pad=4)
plt.show()
# sns.countplot(x='Branch', data=data)
# plt.show()
```

The first chart illustrates the distribution of students across various academic subjects. It reveals that the majority of students have opted for MECH, CIVIL, and IT, based on their respective percentage shares.

The second chart provides a breakdown of student enrollment by gender within each subject. This visualization helps us understand the gender representation and trends across different academic disciplines.

2. Which group of students, across all subjects, secured the highest number of internships and placements?

```python
fig, axes = plt.subplots(1,2, figsize = (10,5))

# subject wise get internship
sns.countplot(x='Branch',
              hue='Internship Done', 
              data=data, 
              palette={'Yes': 'seagreen', 'No': 'firebrick'}, 
              alpha = 0.7, ax= axes[0])

axes[0].set_title("Get an Internship")
axes[0].set_xlabel("Subject")
axes[0].set_ylabel("count")
axes[0].legend(loc='lower right')

# placement based on internship
axes1 = sns.countplot(x = "Branch",
              hue= "Placement Status",
              data= data,
              palette={'Placed':'green', 'Not Placed':'red'},
              alpha = 0.7,
              ax=axes[1])

for p in axes1.patches:
    height = p.get_height()
    axes1.text(p.get_x() + p.get_width() / 2,
             height + 0.5,
             int(height),
             ha='center', va='bottom', fontsize=9)

axes[1].set_title("Placement status")
axes[1].set_xlabel("Subject")
axes[1].set_ylabel("Count")
axes[1].legend(loc='lower right')

plt.tight_layout(pad=4)
plt.show()
```

These two charts compare Placement versus Internship opportunities across different subjects. From the first chart, it's evident that students from MECH, CIVIL, and EEE secured the highest number of internships compared to other disciplines. On the other hand, EEE students achieved the highest number of job placements overall. Interestingly, for CSE students, the number of internships and placements is almost equal, indicating a balanced opportunity in both areas.

3. Placement with anad without internship. 

```python
plt.Figure(figsize=(10,5))
ax = sns.countplot(x = "Internship Done",
              hue= "Placement Status",
              data= data,
              palette={'Placed':'green', 'Not Placed':'coral'},
              alpha = 0.7)

total = len(data)

# Add percentage labels on bars
for p in ax.patches:
    height = p.get_height()
    percent = (height / total) * 100
    ax.text(p.get_x() + p.get_width()/2,
            height + 0.5,
            f'{percent:.1f}%',
            ha='center', va='bottom', fontsize=10)

plt.tight_layout
plt.show()
```

This is the most interesting insight in the data: students who did not complete an internship actually had a higher placement rate than those who did. This suggests that if you possess strong skills and a solid understanding of the job market, an internship may not be a barrier to securing a job. In other words, the right skills and knowledge can outweigh the absence of internship experience.

4. Placement based on 3 skills Data science, Python, ML combine and other seperrate language

```python
# convert my target variable Placement category to numeric
data['Placement_s_num'] = data['Placement Status'].astype('category').cat.codes

# make a new column combine this 3 skill columns
data['python_binary'] = data['python'].map({'Yes':1, 'No': 0})
data['ML_binary'] = data['machine learning'].map({'Yes':1, 'No': 0})
data['data_s_binary'] = data['data science'].map({'Yes':1, 'No': 0})

data['combo(3) skills'] = data['python_binary']+data['ML_binary']+data['data_s_binary']
```

```python
# make all other variable catagory to numeric
data['c++_b'] = data['c++'].map({'Yes':1, 'No': 0}).astype(float)
data['java_binary'] = data['java'].map({'Yes':1, 'No': 0}).astype(float)
data['sql_binary'] = data['sql'].map({'Yes':1, 'No': 0}).astype(float)
data['web development_binary'] = data['web development'].map({'Yes':1, 'No': 0}).astype(float)
```

5. Correlation

```python
# make a sub dataset for correlation
sub_data = data[['Placement_s_num', 'combo(3) skills','c++_b', 'java_binary', 'sql_binary', 'web development_binary']]
```

```python
plt.Figure(figsize=(12, 7))

sns.heatmap(sub_data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

We aimed to examine which individual skills are most correlated with job placement. However, the analysis shows that no single skill has a strong correlation with placement. This suggests that possessing just one skill may not be enough — instead, a combination of multiple skills significantly increases the chances of securing a job.
