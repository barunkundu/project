```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Import csv file

```python
data = pd.read_csv('D:/work hard/Pandas/project/amazon_product/amazon_product.csv')
data
```

```python
data.isna().sum()
```

```python
data.columns
```

Drop unnecessary columns, and also drop some columns that contain nearly all null value

```python
data.drop(columns= ['Unnamed: 0', 'asin', 'currency','product_url', 'product_photo', 'product_num_offers', 'climate_pledge_friendly','delivery', 'has_variations', 'product_availability',
       'unit_price', 'unit_count', 'delivery', 'product_minimum_offer_price','product_num_ratings', 'is_prime'], inplace=True)
```

```python
data.info()
```

Change column name 

```python
data.rename(columns={'is_best_seller': 'best_seller',
             'is_amazon_choice':'amazon_choice',
             'product_price': 'price($)',
             'product_original_price':'product_original_price($)',
             "product_title":'titel'
             }, inplace=True)
data
```

```python
data_new_col = []

for i in data.columns:
    data_new_col.append(i.capitalize())

data_new_col
```

```python
data.columns = data_new_col
```

drop Duplicates

```python
data.drop_duplicates(inplace=True)
```

convert a two column with boolean types "False" and "True" to numeric values 0 and 1

```python
data['Best_seller'] = data['Best_seller'].apply(lambda x: 1 if x == True else 0)
```

```python
data['Amazon_choice'] = data['Amazon_choice'].apply(lambda x: 1 if x == True else 0)
```

```python
data['Sales_volume'] = data['Sales_volume'].str.replace('[^0-9]','', regex=True).replace('',0).astype(float)*100
```

```python
data
```

Null value handling

Replace the "Product_original_price" blank cell with Product_price  

```python
data['Product_original_price($)'] = data['Product_original_price($)'].fillna(data['Price($)'])
data
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data['Price($)']=data['Price($)'].str.replace('$', '')
```

```python
data['Product_original_price($)']=data['Product_original_price($)'].str.replace('$', '')
```

```python
data[['Price($)', 'Product_original_price($)']]=data[['Price($)', 'Product_original_price($)']].astype(float)
data.info()
```

```python
data.boxplot(figsize=(10,8))
plt.xticks(rotation = 45)
plt.tight_layout()
```

round the 'Sales_volume' & 'Product_star_rating'

```python
data_group = data.groupby(['Product_star_rating'])['Sales_volume'].sum()
ax = data_group.plot(kind='line',x='Product_star_rating',y='Sales_volume', xlabel='rating in star', ylabel='Seals volume')
ax.ticklabel_format(style='plain', axis='y')

plt.show()
```

maximum is unique, thats why we are using strip plot

```python
data['price_class'] = pd.cut(data['Price($)'], bins=10)

sns.swarmplot(x='price_class', y='Product_star_rating', data=data)
plt.xticks(rotation = 90)
plt.figure(figsize=(20,8))

plt.tight_layout()
plt.show()
```
